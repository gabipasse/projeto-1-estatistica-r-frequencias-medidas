# **Realize, para a variavel RENDA, uma analise descritiva com as ferramentas que aprendemos em nosso treinamento**

```r
#Media
mean(dados$Renda)
```

```r
#Mediana
median(dados$Renda)
```

```r
Moda <- function(x) {
    freq <- table(x)
    return(names(freq)[freq == max(freq)])
}
```

```r
#Moda
Moda(dados$Renda)
```

```r
#Variancia
var(dados$Renda)
```

```r
variancia.funcao <- function(x){
    media <- mean(x)
    print(media)
    variancia <- sum((x - media) ^ 2) / (length(x) - 1)
    print(variancia)
}
```

```r
variancia.funcao(dados$Renda)
```

```r
#Desvio padrao
sd(dados$Renda)
```

```r
sqrt(var(dados$Renda))
```

### Obtenha a média, mediana e valor máximo da variável RENDA segundo SEXO e COR


```r
#media
medias_renda_cor_sexo <- tapply(dados$Renda, list(dados$Sexo, dados$Cor), mean)

colnames(medias_renda_cor_sexo) <- cor

rownames(medias_renda_cor_sexo) <- sexo

medias_renda_cor_sexo
```

```r
#mediana
medianas_renda_cor_sexo <- tapply(dados$Renda, list(dados$Sexo, dados$Cor), median)

colnames(medianas_renda_cor_sexo) <- cor

rownames(medianas_renda_cor_sexo) <- sexo

medianas_renda_cor_sexo
```

```r
#valor maximo
valor_maximo_renda_cor_sexo <- tapply(dados$Renda, list(dados$Sexo, dados$Cor), max)

colnames(valor_maximo_renda_cor_sexo) <- cor

rownames(valor_maximo_renda_cor_sexo) <- sexo

valor_maximo_renda_cor_sexo
```

```r
#fazer todas as 3 medidas de uma vez <- tenho que fazer de acordo com combinacoes de cor e sexo
funcoes_estatisticas <- list(mean, median, max)

nomes_estatisticas <- c('Media', 'Mediana', 'Maximo')

resultados_funcoes_estatisticas <- lapply(funcoes_estatisticas, function(funcao) {
    
    resultado <- tapply(dados$Renda, list(dados$Sexo, dados$Cor), funcao)
    
    rownames(resultado) <- sexo
    
    colnames(resultado) <- cor
    
    return(resultado)
    
})
                     
resultados_funcoes_estatisticas <- setNames(resultados_funcoes_estatisticas, nomes_estatisticas)    
                     
resultados_funcoes_estatisticas
```

# **Construa um blox plot da variável RENDA segundo SEXO e COR**

```r
library(ggplot2)
```

```r
dados$Cat.Sexo <- factor(dados$Sexo)

dados$Cat.Cor <- factor(dados$Cor)

head(dados, 5)
```

```r
#forma 1
ggplot(data = dados[dados$Renda < 10000, ], aes(x = Cat.Sexo, y = Renda)) +
  geom_boxplot(aes(fill = Cat.Cor)) +
  ylab('R$') +
  xlab('Sexo') +
  ggtitle('Box-plot Renda x Sexo dividido por Cor') +
  facet_grid(Cat.Cor ~ .)
```

```r
formatos <-  theme(
        plot.title = element_text(size = 14, hjust = 0.5),
        axis.title.y = element_text(size= 12, vjust = +0.2),
        axis.title.x = element_text(size= 12, vjust = - 0.2),
        axis.text.y = element_text(size = 15),
        axis.text.x = element_text(size = 15)
    )
```

```r
#forma 2
ggplot(data = dados[dados$Renda < 10000, ], aes(x = Cat.Cor, y = Renda, fill = Cat.Sexo)) +
    geom_boxplot(size = 0.2) +
    coord_flip() +
    ylab('R$') +
    xlab('Cor') +
    guides(fill = guide_legend(title = 'Sexo')) +
    ggtitle('Box plot da RENDA por SEXO e COR') +
    formatos
```
